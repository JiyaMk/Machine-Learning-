```
# Subtask 10

import pandas as pd
import matplotlib.pyplot as plt

a = pd.read_excel("data.xlsx")

month = a['month_number'].tolist()
fc = a['facecream'].tolist()
fw = a['facewash'].tolist()
tp = a['toothpaste'].tolist()
bs = a['bathingsoap'].tolist()
s = a['shampoo'].tolist()
m = a['moisturizer'].tolist()

plt.plot(month,fc,label = "Facecream",color = "pink",linewidth = 4,marker='o')
plt.plot(month,fw,label = "Facewash",color = "blue",linewidth = 4,marker='o')
plt.plot(month,tp,label = "Toothpaste",color = "green",linewidth = 4,marker='o')
plt.plot(month,bs,label = "Bathingsoap",color = "red",linewidth = 4,marker='o')
plt.plot(month,s,label = "Shampoo",color = "purple",linewidth = 4,marker='o')
plt.plot(month,m,label = "Moisturizer",color = "yellow",linewidth = 4,marker='o')

plt.legend(loc='upper left')
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("All Products Sales Data using Multi-line plot")
plt.show
```
